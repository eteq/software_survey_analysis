```python
from __future__ import division

import json

import numpy as np

import astropy.io.ascii
from IPython import display
```

## First we create the relevant json input file

```python
data = astropy.io.ascii.read('questionaire_data.csv', comment='#', delimiter=',')
```

```python
for ii in range(len(data['software'])):
      if data['software'][ii]:
            data['software'][ii] = str(data['software'][ii])
            data['software'][ii] = data['software'][ii].replace('sm or supermongo','sm/supermongo')
            data['software'][ii] = data['software'][ii].replace('Excel or other spreadsheet','Excell/spreadsheet')
            data['software'][ii] = data['software'][ii].replace('Shell Scripting','shell')
            data['software'][ii] = data['software'][ii].replace(' sm-like interface to pgplot. Basically wrote own language ages ago and still cling to it.','')
            data['software'][ii] = data['software'][ii].replace('(which are python-based)','')
            data['software'][ii] = data['software'][ii].replace(' within shell scripting:','')
            data['software'][ii] = data['software'][ii].replace('extension','')
            data['software'][ii] = data['software'][ii].replace('package','')
            data['software'][ii] = data['software'][ii].replace('Marthematica','Mathematica')
            data['software'][ii] = data['software'][ii].replace('II','')
            data['software'][ii] = data['software'][ii].replace('(for data viz & edu tools)','')
            data['software'][ii] = data['software'][ii].replace(' tools','')
            data['software'][ii] = data['software'][ii].replace('-- the ancient "European" IRAF','iraf')
            data['software'][ii] = data['software'][ii].replace('(GNU)','').replace('(USGS)','')
            data['software'][ii] = data['software'][ii].replace('tex','latex')
            data['software'][ii] = data['software'][ii].replace('and','')
            data['software'][ii] = data['software'][ii].replace('.','')
```

```python
software_lists = []
for row in data['software']:
    if row:
        rowentries = []
        entries = row.lower().strip().split(' ')
        for entry in entries:
            if entry:
                if entry in ('c', 'c++'): 
                    entry = 'c/c++'
                rowentries.append(entry)
        software_lists.append(rowentries)
flat_software_lists = np.concatenate(software_lists)
usoftwares = np.unique(flat_software_lists)
software_counts = dict([(nm, np.sum(flat_software_lists==nm)) for nm in usoftwares])
```

```python
for count in range(0,11):
    print 'There are', np.sum(np.array(software_counts.values())>count),'languages with >',count,'users in the survey'
software_w_gtr_unsorted = [(count, nm) for nm, count in software_counts.items() if count > 10]
software_w_gtr = [nm for count,nm in reversed(sorted(software_w_gtr_unsorted))]
software_w_gtr
```

```python
def build_subtree(datalists, langnames):
    if len(datalists) < 1:
        return None
    #note that the input datalists *will* be edited
    #first remove the empty ones, and populate that at the root
    count = 0
    for i, entry in reversed(list(enumerate(datalists))):
        if len(entry) == 0:
            count += 1
            datalists.pop(i)
        
    tree = {'':count}
    
    for lang in langnames:
        langlist = []
        count = 0
        for i, entry in reversed(list(enumerate(datalists))):
            if len(entry) == 0:
                count += 1
                datalists.pop(i)
            elif lang in entry:
                newentry = list(datalists.pop(i))
                newentry.remove(lang)
                langlist.append(newentry)  
        tree[lang] = build_subtree(langlist, langnames)
        if tree[lang] is None:
            del tree[lang]
    return tree

#clean software_lists of anything not in software_w_gtr
cleaned_list = []
for entry in software_lists:
    new_entry = [e for e in entry if e in software_w_gtr]
    if len(new_entry) > 0:
        cleaned_list.append(new_entry)
tree = build_subtree(cleaned_list, software_w_gtr)
```

```python
#now translate into list format suitable for d3
def subtree(nm, dct):
    newdct = {'name': nm}
    newdct['children'] = children = []
    
    dct = dict(dct)
    count = dct.pop('', None)
    if count:
        children.append({'name': nm, 'size': count})
    for langnm in software_w_gtr:
        if langnm in dct:
            children.append(subtree(langnm, dct.pop(langnm)))
    assert len(dct)==0
    return newdct

d3tree = subtree('root', tree)
```

### Sunburst

```python
%%html
<style>

path {
  stroke: #fff;
  fill-rule: evenodd;
}

text {
  font-family: Arial, sans-serif;
  font-size: 9px;
}

div.mousehover {   
  position: absolute;
  top : 50%;
  text-align: center;           
  width: 'auto';                  
  height: 'auto';                 
  padding: 2px;             
  font: 18px sans-serif;        
  background: lightsteelblue;   
  border: 0px;      
  border-radius: 8px;           
  pointer-events: none;
  opacity: 0;
}

</style>

<div class="mousehover"></div>
<div class="d3vizsunburst"></div>

<script>

var width = 1024,
    height = 720,
    radius = Math.min(width, height) / 2;

var x = d3.scale.linear()
    .range([0, 2 * Math.PI]);

var y = d3.scale.linear()
    .range([0, radius]);

var color = d3.scale.category20c();

var svg = d3.select("div.d3vizsunburst").append("svg")
    .attr("width", width)
    .attr("height", height)
  .append("g")
    .attr("transform", "translate(" + width / 2 + "," + (height / 2 + 10) + ")");

var partition = d3.layout.partition()
    .value(function(d) { return d.size; });

var arc = d3.svg.arc()
    .startAngle(function(d) { return Math.max(0, Math.min(2 * Math.PI, x(d.x))); })
    .endAngle(function(d) { return Math.max(0, Math.min(2 * Math.PI, x(d.x + d.dx))); })
    .innerRadius(function(d) { return Math.max(0, y(d.y)); })
    .outerRadius(function(d) { return Math.max(0, y(d.y + d.dy)); });


var g = svg.selectAll("g")
  .data(partition.nodes(window.d3root))
  .enter().append("g");

var mousehover = d3.select("div.mousehover");

var path = g.append("path")
.attr("d", arc)
.style("fill", function(d) { return color((d.children ? d : d.parent).name); })
.on("click", click)
.on("mouseover", function(d) {      
    mousehover.transition()        
        .duration(200)      
        .style("opacity", .9);      
    mousehover.html(get_heirarchy(d))  
        //.style("left", (d3.event.pageX) + "px")     
        //.style("top", (d3.event.pageY - 28) + "px");    
    })                  
.on("mouseout", function(d) {       
    mouseover.transition()        
        .duration(500)      
        .style("opacity", 0);   
});

var text = g.append("text")
.attr("transform", function(d) { return "rotate(" + computeTextRotation(d) + ")"; })
.attr("x", function(d) { return y(d.y); })
.attr("dx", "6") // margin
.attr("dy", ".35em") // vertical-align
.text(function(d) { 
                   if (d.name == 'root') return "";
                   if (typeof d.children === 'undefined') return d.name;
                   return "";
                  });

function click(d) {
// fade out all text elements
text.transition().attr("opacity", 0);

path.transition()
  .duration(750)
  .attrTween("d", arcTween(d))
  .each("end", function(e, i) {
      // check if the animated element's data e lies within the visible angle span given in d
      if (e.x >= d.x && e.x < (d.x + d.dx)) {
        // get a selection of the associated text element
        var arcText = d3.select(this.parentNode).select("text");
        // fade in the text element and recalculate positions
        arcText.transition().duration(750)
          .attr("opacity", 1)
          .attr("transform", function() { return "rotate(" + computeTextRotation(e) + ")" })
          .attr("x", function(d) { return y(d.y); });
      }
  });
}

d3.select(self.frameElement).style("height", height + "px");

// Interpolate the scales!
function arcTween(d) {
  var xd = d3.interpolate(x.domain(), [d.x, d.x + d.dx]),
      yd = d3.interpolate(y.domain(), [d.y, 1]),
      yr = d3.interpolate(y.range(), [d.y ? 20 : 0, radius]);
  return function(d, i) {
    return i
        ? function(t) { return arc(d); }
        : function(t) { x.domain(xd(t)); y.domain(yd(t)).range(yr(t)); return arc(d); };
  };
}

function computeTextRotation(d) {
  return (x(d.x + d.dx / 2) - Math.PI / 2) / Math.PI * 180;
}

function get_heirarchy(d) {
  if (d.name === 'root') return '';
  var texts = [d.name];
  var currnode = d;
  while (currnode.parent.name != 'root') {
    currnode = currnode.parent;
    if (texts[texts.length-1] == currnode.name) {
      texts.splice(0, 0, 'N='+d.size);
    } else {
      texts.push(currnode.name);
    }
  }
  texts.reverse();
  return texts.join('<br>');
  
}
</script>
```

```python
%%html
<style>

circle {
  fill: rgb(31, 119, 180);
  fill-opacity: .25;
  stroke: rgb(31, 119, 180);
  stroke-width: 1px;
}

.leaf circle {
  fill: #ff7f0e;
  fill-opacity: 1;
}

text {
  font: 10px sans-serif;
}

</style>
<div id="d3-viz-circles"></div>

<script>
var diameter = 960,
    format = d3.format(",d");

var pack = d3.layout.pack()
    .size([diameter - 4, diameter - 4])
    .value(function(d) { return d.size; });

var svg = d3.select("#d3-viz-circles").append("svg")
    .attr("width", diameter)
    .attr("height", diameter)
  .append("g")
    .attr("transform", "translate(2,2)");

var node = svg.datum(window.d3root).selectAll(".node")
  .data(pack.nodes)
.enter().append("g")
  .attr("class", function(d) { return d.children ? "node" : "leaf node"; })
  .attr("transform", function(d) { return "translate(" + d.x + "," + d.y + ")"; });

node.append("title")
  .text(function(d) { return d.name + (d.children ? "" : ": " + format(d.size)); });

node.append("circle")
  .attr("r", function(d) { return d.r; });

node.filter(function(d) { return !d.children; }).append("text")
  .attr("dy", ".3em")
  .style("text-anchor", "middle")
  .text(function(d) { return d.name.substring(0, d.r / 3); });


d3.select(self.frameElement).style("height", diameter + "px");
</script>
```
